```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
movies = pd.read_csv('movies.csv')
movies.head(5)
```

```python
movies = movies.drop('genres', axis = 1).join(movies.genres.str.get_dummies())
```

```python
movies
```

```python
frequent_itemset = apriori(movies.drop(['movieId','title','(no genres listed)'],axis = 1), min_support=0.001, use_colnames=True)
```

```python
frequent_itemset
```

```python
m1 = association_rules(frequent_itemset, metric='lift', min_threshold = 1.25)
```

```python
m1
```

```python
association_rules(frequent_itemset)
```

```python
sorted = m1.sort_values(by = ['confidence','lift'], ascending = False)
print(sorted.head())
```

# Clearly, {Crime,IMAX,Action,Drama,Mystery,Thriller,Comedy,Children,Animation,Adventure} is the most associated itemset in this data.
